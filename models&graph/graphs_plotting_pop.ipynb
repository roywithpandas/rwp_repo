```python
#中國東方地區
east_region_pop = ['East-BeijingCity_population_modified.csv',
                   'East-FujianProvince_population_modified.csv',
                   'East-JiangsuProvince_population_modified.csv',
                   'East-ShandongProvince_population_modified.csv',
                   'East-ShanghaiCity_population_modified.csv',
                   'East-ZhejiangProvince_population_modified.csv']

#中國南方地區

south_region_pop = ['South-YunnanProvince_population_modified.csv',
                    'South-GuangdongProvince_population_modified.csv',
                    'South-GuizhouProvince_population_modified.csv',
                    'South-HenanProvince_population_modified.csv',
                    'South-HubeiProvince_population_modified.csv',
                    'South-HunanProvince_population_modified.csv',
                    'South-JiangxiProvince_population_modified.csv',
                    'South-ShaanxiProvince_population_modified.csv']

#中國西方地區

west_region_pop = ['West-ChongqingCity_population_modified.csv',
                   'West-SichuanProvince_population_modified.csv']

```

## Columns listing 

#### Populations : 
- Permanent population at the end of the year (10,000 people) 年末常住人口（萬人）
- Urban population (10,000 people) 城鎮人口（萬人）
- Rural population (10,000 people) 鄉村人口（萬人）
- Birth rate (%) 出生率 （％）
- Population death rate (%) 人口死亡率（%）
- Natural population growth rate (%) 人口自然增長率（%）



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load and concatenate data
def load_and_concat(file_list):
    return pd.concat([pd.read_csv(file) for file in file_list])

# Load and concatenate data for each region
east_pop_data = load_and_concat(east_region_pop)
south_pop_data = load_and_concat(south_region_pop)
west_pop_data = load_and_concat(west_region_pop)

```

### Permanent population at the end of the year (10,000 people) 年末常住人口（萬人）

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average housing prices over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_pop_data, x='Years', y='Permanent population at the end of the year (10,000 people)')
plt.title('East Region Permanent population')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_pop_data, x='Years', y='Permanent population at the end of the year (10,000 people)')
plt.title('South Region Permanent population')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_pop_data, x='Years', y='Permanent population at the end of the year (10,000 people)')
plt.title('West Region Permanent population')

plt.tight_layout()
plt.show()
```

## 人口政策變化 2002 - 2023 

### 二孩政策調整期 ：
    時間
2002 - 2012 ： 雙獨兩孩 
2013 - 2015 ： 單獨兩孩 
2015 - 2021 ： 全面兩孩

### 三孩政策期 ：
    時間 
2021 - now  : 全面三孩 




```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average housing prices over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_pop_data, x='Years', y='Natural population growth rate (%)')
plt.title('East Region population growth rate (%)')

# Add vertical lines at specified years
#'two children alone', 'two children alone', 'two children in total', 'three children in all'
vertical_lines = [2002, 2013, 2015, 2021]
for line in vertical_lines:
    plt.axvline(x=line, color='orange', linestyle='--', label=f'Year {line}')

plt.legend()

plt.subplot(1, 3, 2)
sns.lineplot(data=south_pop_data, x='Years', y='Natural population growth rate (%)')
plt.title('South Region population growth rate (%)')

# Add vertical lines at specified years
for line in vertical_lines:
    plt.axvline(x=line, color='orange', linestyle='--', label=f'Year {line}')

plt.legend()

plt.subplot(1, 3, 3)
sns.lineplot(data=west_pop_data, x='Years', y='Natural population growth rate (%)')
plt.title('West Region population growth rate (%)')

# Add vertical lines at specified years
for line in vertical_lines:
    plt.axvline(x=line, color='orange', linestyle='--', label=f'Year {line}')

plt.legend()

plt.tight_layout()
plt.show()
```
