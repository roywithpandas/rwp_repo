```python
#中國東方地區

east_region_lc = ['East-BeijingCity-landcost_modified.csv',
                  'East-FujianProvince-landcost_modified.csv',
                  'East-JiangsuProvince-landcost_modified.csv',
                  'East-ShandongProvince-landcost_modified.csv',
                  'East-ShanghaiCity-landcost_modified.csv',
                  'East-ZhejiangProvince-landcost_modified.csv']


#中國南方地區

south_region_lc = ['South-GuangdongProvince-landcost_modified.csv',
                   'South-GuizhouProvince-landcost_modified.csv',
                   'South-HenanProvince-landcost_modified.csv',
                   'South-HubeiProvince-landcost_modified.csv',
                   'South-HunanProvince-landcost_modified.csv',
                   'South-JiangxiProvince-landcost_modified.csv',
                   'South-ShaanxiProvince-landcost_modified.csv',
                   'South-YunnanProvince-landcost_modified.csv']

#中國西方地區

west_region_lc = ['West-ChongqingCity-landcost_modified.csv',
                  'West-SichuanProvince-landcost_modified.csv']
```

## Columns listing 

#### land cost:
- Land area to be developed by real estate development enterprises (10k/m2) 房地產開發企業待開發土地面積(萬平方公尺)
- Land area purchased by real estate development companies (10k/m2) 房地產開發企業購置土地面積(萬平方公尺)
- Land transaction price of real estate development enterprises (100million/yuan) 房地產開發企業土地成交價款(億元)
- Land purchase costs for real estate development companies (100million/yuan) 房地產開發企業土地購置費用(億元)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load and concatenate data
def load_and_concat(file_list):
    return pd.concat([pd.read_csv(file) for file in file_list])

# Load and concatenate data for each region
east_lc_data = load_and_concat(east_region_lc)
south_lc_data = load_and_concat(south_region_lc)
west_lc_data = load_and_concat(west_region_lc)

```

### Land area to be developed by real estate development enterprises (10k/m2) 房地產開發企業待開發土地面積(萬平方公尺)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average land cost over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_lc_data, x='Years', y='Land area to be developed by real estate development enterprises (10k/m2)')
plt.title('East Region Land area to be developed')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_lc_data, x='Years', y='Land area to be developed by real estate development enterprises (10k/m2)')
plt.title('South Region Land area to be developed')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_lc_data, x='Years', y='Land area to be developed by real estate development enterprises (10k/m2)')
plt.title('West Region Land area to be developed')

plt.tight_layout()
plt.show()
```

### Land area purchased by real estate development companies (10k/m2) 房地產開發企業購置土地面積(萬平方公尺)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average Land area purchased by real estate development companies over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_lc_data, x='Years', y='Land area purchased by real estate development companies (10k/m2)')
plt.title('East Region Land area purchased (10k/m2)')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_lc_data, x='Years', y='Land area purchased by real estate development companies (10k/m2)')
plt.title('South Region Land area purchased (10k/m2)')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_lc_data, x='Years', y='Land area purchased by real estate development companies (10k/m2)')
plt.title('West Region Land area purchased (10k/m2)')

plt.tight_layout()
plt.show()
```

### Land transaction price of real estate development enterprises (100million/yuan) 房地產開發企業土地成交價款(億元)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average Land transaction price over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_lc_data, x='Years', y='Land transaction price of real estate development enterprises (100million/yuan)')
plt.title('East Region Land transaction price ')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_lc_data, x='Years', y='Land transaction price of real estate development enterprises (100million/yuan)')
plt.title('South Region Land transaction price')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_lc_data, x='Years', y='Land transaction price of real estate development enterprises (100million/yuan)')
plt.title('West Region Land transaction price')

plt.tight_layout()
plt.show()
```

### Land purchase costs for real estate development companies (100million/yuan) 房地產開發企業土地購置費用(億元)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average Land purchase costs over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_lc_data, x='Years', y='Land purchase costs for real estate development companies (100million/yuan)')
plt.title('East Region Land purchase costs')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_lc_data, x='Years', y='Land purchase costs for real estate development companies (100million/yuan)')
plt.title('South Region Land purchase costs')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_lc_data, x='Years', y='Land purchase costs for real estate development companies (100million/yuan)')
plt.title('West Region Land purchase costs')

plt.tight_layout()
plt.show()
```
