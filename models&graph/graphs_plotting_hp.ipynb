```python
#中國東方地區
east_region_hp = ['East-BeijingCity-avghp_modified.csv',
                  'East-FujianProvince-avghp_modified.csv',
                  'East-JiangsuProvince-avghp_modified.csv',
                  'East-ShandongProvince-avghp_modified.csv',
                  'East-ShanghaiCity-avghp_modified.csv',
                  'East-ZhejiangProvince-avghp_modified.csv']

east_region_pop = ['East-BeijingCity_population_modified.csv',
                   'East-FujianProvince_population_modified.csv',
                   'East-JiangsuProvince_population_modified.csv',
                   'East-ShandongProvince_population_modified.csv',
                   'East-ShanghaiCity_population_modified.csv',
                   'East-ZhejiangProvince_population_modified.csv']

#中國南方地區
south_region_hp =['South-YunnanProvince-avghp_modified.csv',
                  'South-GuangdongProvince-avghp_modified.csv',
                  'South-GuizhouProvince-avghp_modified.csv',
                  'South-HenanProvince-avghp_modified.csv',
                  'South-HubeiProvince-avghp_modified.csv',
                  'South-HunanProvince-avghp_modified.csv',
                  'South-JiangxiProvince-avghp_modified.csv',
                  'South-ShaanxiProvince-avghp_modified.csv']

south_region_pop = ['South-YunnanProvince_population_modified.csv',
                    'South-GuangdongProvince_population_modified.csv',
                    'South-GuizhouProvince_population_modified.csv',
                    'South-HenanProvince_population_modified.csv',
                    'South-HubeiProvince_population_modified.csv',
                    'South-HunanProvince_population_modified.csv',
                    'South-JiangxiProvince_population_modified.csv',
                    'South-ShaanxiProvince_population_modified.csv']

#中國西方地區

west_region_hp = ['West-ChongqingCity-avghp_modified.csv',
                  'West-SichuanProvince-avghp_modified.csv']

west_region_pop = ['West-ChongqingCity_population_modified.csv',
                   'West-SichuanProvince_population_modified.csv']

```

## Columns listing 

#### Housing Price : 
- Average sales price of commercial housing (yuan/m^2) 商品住宅平均銷售價格（元/平方米） Y 
- Average sales price of residential commercial buildings (yuan/m^2) 住宅商品房平均銷售價格（元/平方米） Y 
- Average sales price of villas and high-end apartments (yuan/m^2) 別墅及高級公寓平均銷售價格（元/平方米） Y 
- Average sales price of commercial buildings in office buildings (yuan/m^2) 辦公室商品房平均銷售價格（元/平方米）
- Average sales price of commercial buildings (yuan/m^2) 商品房平均銷售價格（元/平方米）Y 
- Average sales price of other commercial buildings (yuan/m^2) 其他商業建築平均銷售價格（元/平方米）


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load and concatenate data
def load_and_concat(file_list):
    return pd.concat([pd.read_csv(file) for file in file_list])

# Load and concatenate data for each region
east_hp_data = load_and_concat(east_region_hp)
south_hp_data = load_and_concat(south_region_hp)
west_hp_data = load_and_concat(west_region_hp)

```

### Average sales price of commercial housing (yuan/m^2) 商品住宅平均銷售價格（元/平方米）

```python
# Set up the plotting style
sns.set(style="ticks")

# Plotting average housing prices over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_hp_data, x='Years', y='Average sales price of commercial housing (yuan/m^2)')
plt.title('East Region commercial housing')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_hp_data, x='Years', y='Average sales price of commercial housing (yuan/m^2)')
plt.title('South Region commercial housing')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_hp_data, x='Years', y='Average sales price of commercial housing (yuan/m^2)')
plt.title('West Region commercial housing')

plt.tight_layout()
plt.show()
```

### Average sales price of residential commercial buildings (yuan/m^2)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average housing prices over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_hp_data, x='Years', y='Average sales price of residential commercial buildings (yuan/m^2)')
plt.title('East Region residential commercial ')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_hp_data, x='Years', y='Average sales price of residential commercial buildings (yuan/m^2)')
plt.title('South Region residential commercial ')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_hp_data, x='Years', y='Average sales price of residential commercial buildings (yuan/m^2)')
plt.title('West Region residential commercial ')

plt.tight_layout()
plt.show()
```

### Average sales price of villas and high-end apartments (yuan/m^2) 
別墅及高級公寓平均銷售價格（元/平方米)

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Plotting average housing prices over time for each region
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(data=east_hp_data, x='Years', y='Average sales price of villas and high-end apartments (yuan/m^2)')
plt.title('East Region villas and high-end apartments')

plt.subplot(1, 3, 2)
sns.lineplot(data=south_hp_data, x='Years', y='Average sales price of villas and high-end apartments (yuan/m^2)')
plt.title('South Region villas and high-end apartments')

plt.subplot(1, 3, 3)
sns.lineplot(data=west_hp_data, x='Years', y='Average sales price of villas and high-end apartments (yuan/m^2)')
plt.title('West Region villas and high-end apartments')

plt.tight_layout()
plt.show()
```

### Average sales price of other commercial buildings (yuan/m^2) 其他商業建築平均銷售價格（元/平方米）
