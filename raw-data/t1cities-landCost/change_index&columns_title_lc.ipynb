```python
import pandas as pd
import os

# List of eastern provinces' names
province_names = ['East-BeijingCity', 'East-FujianProvince','East-JiangsuProvince',
                  'East-ShandongProvince', 'East-ShanghaiCity', 'East-ZhejiangProvince',
                  'South-GuangdongProvince','South-GuizhouProvince','South-HenanProvince',
                  'South-HubeiProvince','South-HunanProvince','South-JiangxiProvince',
                  'South-ShaanxiProvince','South-YunnanProvince','West-ChongqingCity',
                  'West-SichuanProvince']

# Loop through each province
for province in province_names:
    file_path = f'{province}-landcost.csv'

    # Read the CSV file
    df = pd.read_csv(file_path)

    # Convert the "column1" to datetime format with a custom format and then extract the year
    df['时间'] = pd.to_datetime(df['时间'], format='%Y年').dt.year

    # Set "column1" as the new index
    df.set_index('时间', inplace=True)

    # Save the modified DataFrame back to the CSV file
    df.to_csv(f'{province}-landcost.csv', index_label='Years')

# Display the first few rows of one of the modified DataFrames
print(df.head())
```

```python
import pandas as pd
import os

# List of file names
file_names = ['East-BeijingCity-landcost.csv', 'East-FujianProvince-landcost.csv', 'East-JiangsuProvince-landcost.csv',
              'East-ShandongProvince-landcost.csv', 'East-ShanghaiCity-landcost.csv', 'East-ZhejiangProvince-landcost.csv',
              'South-GuangdongProvince-landcost.csv', 'South-GuizhouProvince-landcost.csv', 'South-HenanProvince-landcost.csv',
              'South-HubeiProvince-landcost.csv', 'South-HunanProvince-landcost.csv', 'South-JiangxiProvince-landcost.csv',
              'South-ShaanxiProvince-landcost.csv', 'South-YunnanProvince-landcost.csv', 'West-ChongqingCity-landcost.csv',
              'West-SichuanProvince-landcost.csv']

# Columns mapping
column_mapping = {
    '房地产开发企业待开发土地面积(万平方米)': 'Land area to be developed by real estate development enterprises (10k/m2)',
    '房地产开发企业购置土地面积(万平方米)': 'Land area purchased by real estate development companies (10k/m2)',
    '房地产开发企业土地成交价款(亿元)': 'Land transaction price of real estate development enterprises (100million/yuan)',
    '房地产开发企业土地购置费用(亿元)': 'Land purchase costs for real estate development companies (100million/yuan)'}

# Specify the directory path
output_directory = 'landCost_modified'

# Create the directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through each file
for file_name in file_names:
    # Read the CSV file
    df = pd.read_csv(file_name)

    # Rename columns
    df.rename(columns=column_mapping, inplace=True)

    # Extract the region and province names from the file name
    parts = file_name.split('-')
    region_name = parts[0]
    province_name = parts[1].split('_')[0]

    # Save the modified DataFrame back to the CSV file in the specified directory
    output_file_path = os.path.join(output_directory, f'{region_name}-{province_name}-landcost_modified.csv')
    df.to_csv(output_file_path, index=False)

    # Display the first few rows of the modified DataFrame
    print(f"Modified DataFrame for {region_name}-{province_name}:")
    print(df.head())
    print("\n")
```
