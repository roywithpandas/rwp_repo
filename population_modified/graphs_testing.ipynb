```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names
file_names = ['East-BeijingCity_population_modified.csv',
              'East-FujianProvince_population_modified.csv',
              'East-JiangsuProvince_population_modified.csv',
              'East-ShandongProvince_population_modified.csv',
              'East-ShanghaiCity_population_modified.csv',
              'East-ZhejiangProvince_population_modified.csv',
              'South-GuangdongProvince_population_modified.csv',
              'South-GuizhouProvince_population_modified.csv',
              'South-HenanProvince_population_modified.csv',
              'South-HubeiProvince_population_modified.csv',
              'South-HunanProvince_population_modified.csv',
              'South-JiangxiProvince_population_modified.csv',
              'South-ShaanxiProvince_population_modified.csv',
              'South-YunnanProvince_population_modified.csv',
              'West-ChongqingCity_population_modified.csv',
              'West-SichuanProvince_population_modified.csv']

# Create an empty DataFrame to store the data
population_data = pd.DataFrame()

# Loop through each file
for file_name in file_names:
    # Read the CSV file
    df = pd.read_csv(file_name, index_col='Years', parse_dates=['Years'])  # Assuming 'Year' is the index

    # Extract the region and city from the file name
    region_parts = file_name.split('-')

    # Handle cases where the file name structure is different
    if len(region_parts) > 2:
        region = region_parts[1]
        city = region_parts[2].split('_')[0].replace('_population_modified.csv', '')
    else:
        region = region_parts[0]
        city = region_parts[1].replace('_population_modified.csv', '')

    # Extract the 'Permanent population at the end of the year (10,000 people)' column
    df_selected = df[['Permanent population at the end of the year (10,000 people)']]

    # Rename the column to include the region and city
    df_selected.rename(columns={'Permanent population at the end of the year (10,000 people)': f'Population ({region} - {city})'}, inplace=True)

    # Merge the data into the main DataFrame with a specified suffix
    population_data = pd.merge(population_data, df_selected, left_index=True, right_index=True, how='outer', suffixes=('', f'_{region}_{city}'))

# Convert the index to datetime
population_data.index = pd.to_datetime(population_data.index)

# Plot all regions on one chart with customized legend labels
plt.figure(figsize=(16, 10))

for column in population_data.columns:
    plt.plot(population_data.index, population_data[column], marker='o', label=column)

plt.title('Permanent Population Change (2002-2022) - All Regions')
plt.xlabel('Year')
plt.ylabel('Permanent Population (10,000 people)')

# Customize legend labels
legend_labels = [column.split('_')[1] if '_' in column else column for column in population_data.columns]
plt.legend(title='Region - City', labels=legend_labels, loc='upper left')

plt.grid(True)
plt.show()
```
